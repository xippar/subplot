## ข้อมูลที่ต้องใช้

#### นำเข้า pandas

```python
import pandas as pd
```

#### เปิดข้อมูลจากไฟล์ North.csv

```python
table = pd.read_csv('North.csv',index_col=[0])
quarter_sales = table.transpose()
quarter_sales
```

## สร้างกราฟ

#### นำเข้า matplotlib

```python
import matplotlib.pyplot as plt
```

#### สร้าง figure เติม subplot (พร้อม save กราฟเป็นไฟล์ .png)

```python
fig, axes = plt.subplots(1,2,figsize=(14,4))
quarter_sales.plot(ax=axes[0])
axes[0].legend(bbox_to_anchor=(-0.2,1))

quarter_sales.plot.bar(ax=axes[1])
axes[1].legend(bbox_to_anchor=(1,1))
plt.tight_layout()
plt.savefig('subplots.png')
```
